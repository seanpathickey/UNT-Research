## Objective: Loop through folder of PA sensor files and average hourly
#### Loop through folder
#### Import file as dataframe
#### Select only April-September for df in each file (filter)
#### Convert to datetime format
#### Drop minutes and seconds from timestamp
#### Groupby date at the hour resolution 
#### Output to a new file for each sensor



```python
import pandas as pd                 # pandas allows me to read and manipulate my csv data with python in dataframes
from datetime import datetime       #datetime allows me to parse through datetime timestamps in cells and isolate specific parts of the timestamp (e.g. hour, minute sec)
import os                           #os allows me to run the code on each file within a folder

directory = os.listdir(r"C:\Users\sph0088\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II-SD\Combined")
os.chdir(r"C:\Users\sph0088\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II-SD\Combined")

count = 0
for file in directory:
    df = pd.read_csv(file)
    df = df.loc[(df['UTCDateTime'] > '2020/03/31') & (df['UTCDateTime'] <= '2020/10/01')] #filter dataframe by date
    df['UTCDateTime'] = pd.to_datetime(df['UTCDateTime'], format="%Y/%m/%dT%H:%M:%Sz") #convert to datetime
#     df = df[['UTCDateTime', 'mac_address',
#        'current_temp_f', 'current_humidity', 'current_dewpoint_f', 'pressure',
#        'pm1_0_atm', 'pm2_5_atm', 'pm10_0_atm',
#        'pm1_0_atm_b', 'pm2_5_atm_b', 'pm10_0_atm_b']]
    #Average hourly
    df['UTCDateTime'] = df['UTCDateTime'].dt.floor('H') #Drops off the minutes and seconds of the time
    df = pd.DataFrame(df.groupby(['UTCDateTime'],as_index=False).mean()) #groupby hour
    filename = os.path.splitext(file)[0] #split extension from filename
    df.to_csv(r'C:\Users\sph0088\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II_PA-II-SD_new\{}.csv'.format(filename),index=False)
    count = count + 1
    print(df.head())


```
